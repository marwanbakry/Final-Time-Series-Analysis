```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.tsa.arima_model import ARIMA
import numpy as np

def check_stationarity(timeseries):
    adf_test = adfuller(timeseries)
    critical_values = adf_test[4]

    print(f"ADF Statistic: {adf_test[0]:.4f}")
    print(f"p-value: {adf_test[1]:.4f}")
    print(f"Critical Values:")
    for key, value in critical_values.items():
        print(f"{key} ({value:.4f})")

    is_stationary = adf_test[1] < 0.05
    return is_stationary

def plot_acf_pacf(timeseries):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

    acf_values = acf(timeseries)
    pacf_values = pacf(timeseries)

    lags = np.arange(len(acf_values))

    ax1.bar(lags, acf_values, color='skyblue', alpha=0.7)
    ax1.axhline(y=0, color='r', linestyle='--', alpha=0.7)
    ax1.axhline(y=1.96/np.sqrt(len(timeseries)), color='gray', linestyle='--', alpha=0.7)
    ax1.axhline(y=-1.96/np.sqrt(len(timeseries)), color='gray', linestyle='--', alpha=0.7)
    ax1.set_title('Autocorrelation Function')
    ax1.set_xlabel('Lags')  #
    ax1.set_ylabel('ACF')
    ax1.set_xticks(lags)

    ax2.bar(lags, pacf_values, color='skyblue', alpha=0.7)
    ax2.axhline(y=0, color='r', linestyle='--', alpha=0.7)
    ax2.axhline(y=1.96/np.sqrt(len(timeseries)), color='gray', linestyle='--', alpha=0.7)
    ax2.axhline(y=-1.96/np.sqrt(len(timeseries)), color='gray', linestyle='--', alpha=0.7)
    ax2.set_title('Partial Autocorrelation Function')
    ax2.set_xlabel('Lags')
    ax2.set_ylabel('PACF')
    ax2.set_xticks(lags)
    plt.tight_layout()
    plt.show()

df = pd.read_csv('Amazon stock 1999 to 2022.csv')

df['Date'] = pd.to_datetime(df['Date'])

df.set_index('Date', inplace=True)

timeseries_data = df['High']

print("Checking stationarity of the original time series:")
is_stationary = check_stationarity(timeseries_data)

if not is_stationary:
    differenced_data = timeseries_data.diff().dropna()

    print("\nChecking stationarity of the differenced time series:")
    is_stationary_after_diff = check_stationarity(differenced_data)

    print("\nPlotting ACF and PACF of the differenced time series:")
    plot_acf_pacf(differenced_data)

    if is_stationary_after_diff:
        print("\nThe differenced series is now stationary.")
    else:
        print("\nFurther preprocessing may be needed to achieve stationarity.")

else:
    print("\nThe original time series is stationary.")
```

```python
train_data = df['High'][:len(df) // 2]
test_data = df['High'][len(df) // 2:]
```

```python
p, d, q = 4, 2, 1
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train_data, order=(p, d, q))
fitted_model = model.fit()
```

```python
start_idx = len(train_data)
end_idx = len(df) - 1
predictions = fitted_model.predict(start=start_idx, end=end_idx, typ='levels')
```

```python
predictions = fitted_model.predict(start=len(train_data), end=len(df) - 1)
predictions
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Calculate evaluation metrics
mse = mean_squared_error(test_data, predictions)
rmse = np.sqrt(mse)
mae = mean_absolute_error(test_data, predictions)

absolute_errors = np.abs(test_data - predictions)
percentage_errors = absolute_errors / test_data.replace(0, np.nan)
percentage_errors = percentage_errors.fillna(0)
mape = np.mean(percentage_errors) * 100

print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['High'], label='Actual High Prices', color='blue')
plt.plot(test_data.index, predictions, label='Predicted High Prices', color='red')
plt.title('ARIMA Model Forecasting')
plt.xlabel('Date')
plt.ylabel('High Prices')
plt.legend()
plt.grid(True)
plt.show()
```

```python
size = int(len(df) - 30)
train = df['High'][:size]
test = df['High'][size:]

print('\t SARIMA MODEL : In - Sample Forecasting \n')
history = [x for x in train]
predictions = []
```

```python
max_forecast_steps = 100

for t in range(min(len(test), max_forecast_steps)):
    model = SARIMAX(history, order=(2, 1, 2), seasonal_order=(0, 1, 1, 12))
    model_fit = model.fit()

    output = model_fit.forecast()
    yhat = output[0]

    predictions.append(float(yhat))

    obs = test[t]
    history.append(obs)

    print(f'Predicted = {yhat:.2f}, Expected = {obs:.2f}')

predictions_series = pd.Series(predictions, index=test.index[:len(predictions)])
```

```python
# Plotting the predicted values
plt.figure(figsize=(15, 5))
plt.plot(test.index[:len(predictions)], predictions_series, label='Predicted Values', color='red')
plt.plot(test, label='Expected Values', color='blue')
plt.title('SARIMA Model Forecasting for Amazon Stock Prices')
plt.xlabel('Date')
plt.ylabel('High Prices')
plt.legend()
plt.grid(True)
plt.show()
```

```python
closing_prices = df['Close']

test_size = 30
train_prices = closing_prices[:-test_size]
test_prices = closing_prices[-test_size:]

order = (2, 1, 2)           # Non-seasonal (p, d, q)
seasonal_order = (0, 1, 1, 12)  # Seasonal (P, D, Q, S)

model = SARIMAX(train_prices, order=order, seasonal_order=seasonal_order)

model_fit = model.fit()

forecast_steps = len(test_prices)
forecast = model_fit.forecast(steps=forecast_steps)

rmse = np.sqrt(mean_squared_error(test_prices, forecast))
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")

plt.figure(figsize=(14, 8))

# Plotting the train data
plt.plot(train_prices.index, train_prices, label='Train Data', color='blue')

# Plotting the test data (actual closing prices)
plt.plot(test_prices.index, test_prices, label='Test Data', color='green')

# Plotting the forecasted closing prices
plt.plot(test_prices.index, forecast, label='Forecasted Prices', color='red')

plt.title('SARIMA Forecasting for Closing Prices')
plt.xlabel('Date')
plt.ylabel('Closing Prices')
plt.legend()

plt.grid(True)

plt.show()
```

```python

# Get the differenced series (first difference)
diff_series = closing_prices.diff(periods=1)

# Integrated (differenced) series represents the trend component
trend_component = diff_series.cumsum()

# Plot the trend component
plt.figure(figsize=(12, 6))
plt.plot(closing_prices.index, closing_prices, label='Original Data', color='blue', alpha=0.5)
plt.plot(trend_component.index, trend_component, label='Trend Component', color='red')
plt.title('Trend Component of Closing Prices (SARIMA)')
plt.xlabel('Date')
plt.ylabel('Closing Prices')
plt.legend()
plt.grid(True)
plt.show()
```
